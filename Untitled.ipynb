```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Step 1: Load your dataset from a CSV file
data = pd.read_csv('new_dataset_filled.csv')

# Step 2: Data Cleaning
# Replace non-numeric values (e.g., '-') with NaN
data = data.replace('-', np.nan)

# Remove rows with missing values in the relevant columns
data = data.dropna(subset=['Module', 'Product Type', 'Total SMV', 'Avg Per Day', 'Actual Per Day', 'Defect Qty', 'Reject Qty', 'Missing Qty'])

# Step 3: Prepare the data for modeling
X = data[['Product Type', 'Total SMV', 'Avg Per Day', 'Actual Per Day', 'Defect Qty', 'Reject Qty', 'Missing Qty']]
y = data['Module']

# Step 4: Encode the 'Module' and 'Product Type' labels
module_label_encoder = LabelEncoder()
product_type_label_encoder = LabelEncoder()

y_encoded = module_label_encoder.fit_transform(y)
X['Product Type'] = product_type_label_encoder.fit_transform(X['Product Type'])

# Step 5: Feature Engineering
X['Defect_Rate'] = X['Defect Qty'] / X['Total SMV']
X['Avg_Diff'] = X['Actual Per Day'] - X['Avg Per Day']

# Create a new feature 'Avg_Per_Day_Diff_From_Mean'
X['Avg_Per_Day_Mean'] = data.groupby('Product Type')['Avg Per Day'].transform('mean')
X['Avg_Per_Day_Diff_From_Mean'] = X['Avg Per Day'] - X['Avg_Per_Day_Mean']

# Step 6: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Step 7: Sequence Data Preparation
# You need to create sequences and pad them to have a fixed length.
sequence_length = 10  # Adjust this to the desired sequence length
X_train_sequence = [X_train[X_train['Product Type'] == product_type].values for product_type in X['Product Type'].unique()]
X_test_sequence = [X_test[X_test['Product Type'] == product_type].values for product_type in X['Product Type'].unique()]

# Padding sequences
X_train_sequence = pad_sequences(X_train_sequence, maxlen=sequence_length, dtype='float32', padding='post')
X_test_sequence = pad_sequences(X_test_sequence, maxlen=sequence_length, dtype='float32', padding='post')

# Step 8: Build an RNN model
model = Sequential()
model.add(Embedding(input_dim=len(X['Product Type'].unique()), output_dim=8, input_length=sequence_length))
model.add(LSTM(64))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(len(module_label_encoder.classes_), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 9: Train the RNN model
model.fit(X_train_sequence, y_train, epochs=10, batch_size=64, validation_data=(X_test_sequence, y_test))

# Step 10: Make predictions for a given Product Type
product_type = 'Tank Top'  # Replace with your desired product type
product_type_encoded = product_type_label_encoder.transform([product_type])

# Create a sequence for the given product type
product_type_data = X[X['Product Type'] == product_type].values
product_type_sequence = pad_sequences([product_type_data], maxlen=sequence_length, dtype='float32', padding='post')

predicted_module_encoded = model.predict_classes(product_type_sequence)
predicted_module = module_label_encoder.inverse_transform(predicted_module_encoded)

# Print the predicted module
print(f"Best Module for Product Type '{product_type}': {predicted_module[0]}")

# Step 11: Evaluate the RNN model
test_loss, test_accuracy = model.evaluate(X_test_sequence, y_test)
print(f"Test accuracy: {test_accuracy * 100:.2f}%")
```
